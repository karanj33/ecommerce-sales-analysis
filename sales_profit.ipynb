# converting colums into date and time

```python
data['Order Date']=pd.to_datetime(data['Order Date'])
data['Ship Date']=pd.to_datetime(data['Ship Date'])
```

```python
data.head(2)
```

# separating month and year calculating shipping days

```python
data['Order Month']=data['Order Date'].dt.month
data['Order Year']=data['Order Date'].dt.year
data["Delivery Days"] = (data["Ship Date"] - data["Order Date"]).dt.days
```

# Business Case Problem

```python
total_profit = data['Profit'].sum()
print(total_profit)
```

```python
total_quantity = data['Quantity'].sum()
print(total_quantity)
```

```python
total_sales = data['Sales'].sum()
print(total_sales)
```

```python
segment_summary = data.groupby('Segment')[['Sales', 'Profit']].sum().reset_index()
print(segment_summary)
```

```python
category_sales = data.groupby('Category')['Sales'].sum().reset_index()
print(category_sales)
```

```python
subcategory_sales = data.groupby('Sub-Category')['Sales'].sum().reset_index().sort_values(by='Sales', ascending=False)
print(subcategory_sales)
```

```python
monthly_sales = data.groupby('Order Month')['Sales'].sum().reset_index().sort_values(by='Order Month')
print(monthly_sales)
```
